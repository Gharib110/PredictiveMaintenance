```python
merged_reviews.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows with NaN values
merged_reviews.dropna(inplace=True)
```

```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters and digits
    text = re.sub(r'[^a-zA-Z]', ' ', text)

    # Tokenize the text
    tokens = nltk.word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join the tokens back into a single string
    clean_text = ' '.join(tokens)

    return clean_text
```

```python
merged_reviews['clean_text'] = merged_reviews['review_text'].apply(preprocess_text)
```

```python
merged_reviews['clean_text'] = merged_reviews['review_text'].apply(preprocess_text)
```

```python
vectorizer = TfidfVectorizer(min_df=1, max_features=6000, ngram_range=(1, 2))  # Set min_df to 1 and add default_word to
X_features = vectorizer.fit_transform(merged_reviews['clean_text'])
```

```python
# Define the number of folds for cross-validation
num_folds = 5

# Perform k-fold cross-validation
kfold = KFold(n_splits=num_folds)
accuracies = []
y = np.array(merged_reviews['review_score'])
for train_idx, test_idx in kfold.split(X_features):
    X_train, X_test = X_features[train_idx], X_features[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    # Define the LogisticRegression classifier with your chosen hyperparameters
    lg = LogisticRegression(solver='newton-cholesky', max_iter=500, multi_class='ovr', n_jobs=-1)

    # Fit the lg model on the training data
    lg.fit(X_train, y_train)

    # Predict on the test data
    y_pred = lg.predict(X_test)

    # Calculate accuracy for the fold
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Calculate the average accuracy across all folds
average_accuracy = sum(accuracies) / num_folds

# Print the average accuracy
print("Average Accuracy:", average_accuracy)
```
